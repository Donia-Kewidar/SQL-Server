# Seaborn Library
-----------------------------------------------------------------------

Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for creating informative and attractive statistical graphics. Let's go through three commonly used Seaborn functions: relplot, displot, and catplot, along with their various types and examples.

## relplot (Relational Plot)
relplot is used for creating scatter plots and other relational plots. It's a high-level interface that combines the functionality of scatterplot and lineplot. It's particularly useful for visualizing the relationship between two numeric variables.

### 1-Scatter plot
A scatter plot is used to visualize the relationship between two numerical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")
tips
# Scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips
```

###  2-Line Plot
A line plot is useful for visualizing the trend in a numerical variable over time or another continuous variable.

```python
# Line plot
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line Plot: Total Bill vs Tip")
plt.show()
```

## catplot (Categorical Plot):
catplot is used for visualizing the relationship between one numeric and one or more categorical variables. It's a generalization of the relplot function and supports various plot types.

### 1- Bar Plot
A bar plot is suitable for visualizing the distribution of a categorical variable.

```python
# Bar plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Bar Plot: Total Bill by Day")
plt.show()


# OR
# # Bar Plot
# sns.catplot(x="day", y="total_bill", data=tips, kind="bar")
# plt.title("Bar Plot of Average Total Bill by Day")
# plt.show()
```

### 2- Box Plot
A box plot is useful for displaying the distribution of a numerical variable and identifying outliers.

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot: Total Bill by Day")
plt.show()


# OR
# # Box Plot
# sns.catplot(x="day", y="total_bill", data=tips, kind="box")
# plt.title("Box Plot of Total Bill by Day")
# plt.show()
```

### 3- Violin Plot
A violin plot combines aspects of a box plot and a kernel density plot to show the distribution of a variable.

```python
# Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot: Total Bill by Day")
plt.show()


# OR

# # Violin Plot
# sns.catplot(x="day", y="total_bill", data=tips, kind="violin")
# plt.title("Violin Plot of Total Bill by Day")
# plt.show()
```

### 4- count plot
A count plot is used to show the counts of observations in cat

```python
# Count plot
sns.countplot(x="day", data=tips)
plt.title("Count Plot: Count of Observations by Day")
plt.show()

# OR
# # Count Plot
# sns.catplot(x="day", data=tips, kind="count", palette="viridis")
# plt.title("Count Plot of Days")
# plt.show()
```

### 5- Swarm Plot
A swarm plot is useful for visualizing the distribution of a categorical variable with respect to a numerical variable.

```python
# Swarm plot
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Swarm Plot: Total Bill by Day")
plt.show()


# OR
# # Swarm Plot
# sns.catplot(x="day", y="total_bill", data=tips, kind="swarm")
# plt.title("Swarm Plot of Total Bill by Day")
# plt.show()
```

### 6- Strip Plot
A strip plot is useful for visualizing the distribution of a categorical variable against a numerical variable.

```python
# Strip plot
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("Strip Plot: Total Bill by Day")

# OR
# # Strip Plot
# sns.catplot(x="day", y="total_bill", data=tips, kind="strip")
# plt.title("Strip Plot of Total Bill by Day")
# plt.show()
```

### 7- Point Plot
A point plot is similar to a bar plot but represents the estimate of central tendency with points.

```python
# Point plot with markeredgewidth
sns.pointplot(x="day", y="total_bill", data=tips,  markers=["o", "x"], linestyles=["-", "--"], markeredgewidth=1.5)
plt.title("Point Plot: Total Bill by Day with Hue")
plt.show()


# OR
# # Point Plot
# sns.catplot(x="day", y="total_bill", data=tips, kind="point")
# plt.title("Point Plot of Total Bill by Day")
# plt.show()
```

### 8- Boxen Plot
is a type of categorical plot in Seaborn that is an extension of the box plot. It is designed to provide more information about the shape of the distribution, especially in the tails. Boxen plots are useful when you want to visualize the distribution of a variable across different categories and gain insights into the shape of the data distribution.

```python
# Boxen Plot
sns.catplot(x="day", y="total_bill", data=tips, kind="boxen")
plt.title("Boxen Plot of Total Bill by Day")
plt.show()
```

## displot (Distribution Plot)
displot is used for visualizing univariate distributions of a dataset. It allows you to examine the distribution of a single variable.

### 1- Histogram
A histogram is used to visualize the distribution of a single numerical variable.

```python
# Histogram
# sns.histplot(tips['total_bill'], bins=20, kde=True)
# plt.title("Histogram: Distribution of Total Bill")
# plt.show()


# OR
# Histogram
sns.displot(tips["total_bill"], kde=True, bins=20)
plt.title("Histogram of Total Bill")
plt.show()
```

### 2- Displot
using displot to visualize the distribution of a univariate dataset

```python
# Distplot
sns.displot(tips['total_bill'])
plt.title("Displot: Distribution of Total Bill")
plt.show()
```

### 3-Rug Plot
A rug plot is used to visualize the distribution of a single numerical variable along the x-axis or y-axis.

```python
# Rug plot
sns.rugplot(x="total_bill", data=tips, height=0.5)
plt.title("Rug Plot: Distribution of Total Bill")
plt.show()
```

### 4-KDE Plot
A KDE (Kernel Density Estimate) plot visualizes the distribution of a single variable as a smoothed curve.

```python
# KDE plot
sns.kdeplot(tips['total_bill'], shade=True)
plt.title("KDE Plot: Distribution of Total Bill")
plt.show()


# OR
# # KDE Plot
# sns.displot(tips["total_bill"], kind="kde")
# plt.title("Kernel Density Estimate of Total Bill")
# plt.show()
```

### 5- ecdfplot
Plots the empirical cumulative distribution function

```python
# ECDF Plot
sns.displot(tips["total_bill"], kind="ecdf")
plt.title("Empirical Cumulative Distribution Function of Total Bill")
plt.show()
```

##  Pair Plot
A pair plot is used to visualize pairwise relationships in a dataset.

```python
# Pair plot
sns.pairplot(tips, hue="sex")
plt.title("Pair Plot: Pairwise Relationships with Hue")
plt.show()
```

## PairGrid
PairGrid is a versatile grid for displaying pairwise relationships in a dataset. It allows you to customize the type of plots along the diagonal and off-diagonal.

```python
# PairGrid
g = sns.PairGrid(tips, hue="sex")
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot)  # Remove cmap parameter
plt.suptitle("PairGrid: Pairwise Relationships with Hue")
plt.show()
```

## Pairwise Relationships with hue and style
Customizing the hue and style in a Pair Plot for more insights

```python
# Pair Plot with hue and style
sns.pairplot(tips, hue="sex", markers=["o", "s"], palette="husl", diag_kind="kde")
plt.suptitle("Pair Plot: Pairwise Relationships with Hue and Style")
plt.show()
```

## Pairwise Relationships with Regression Lines
Enhancing Pair Plot with regression lines and kernel density estimates.

```python
# Pair Plot with regression lines and KDE
sns.pairplot(tips, kind="reg", diag_kind="kde")
plt.suptitle("Pair Plot: Pairwise Relationships with Regression Lines and KDE")
plt.show()
```

## Pairwise Relationships with Kernel Density and Hexbin
Combining kernel density estimates and hexbin plots in a pair plot.

```python
# Pair Plot with Kernel Density and Hexbin
sns.pairplot(tips, kind="scatter", diag_kind="kde", hue="sex", markers=["o", "s"], palette="Set1", plot_kws={"alpha": 0.5})
plt.suptitle("Pair Plot with KDE and Hexbin: Pairwise Relationships with Hue")
plt.show()
```

## Joint Plot
A joint plot combines multiple plots for bivariate data analysis, including scatter plots and histograms.

```python
# Joint plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.suptitle("Joint Plot: Total Bill vs Tip")
plt.show()
```

## Pairwise Density Plots
Pairwise density plots can be used to visualize the bivariate distribution of two variables.

```python
# Pairwise Density Plots
sns.set(style="white")
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", color="m")
plt.suptitle("Pairwise Density Plots: Total Bill vs Tip")
plt.show()
```

## Hexbin Plot
A hexbin plot is useful for visualizing the distribution of two variables with hexagonal bins.

```python
# Hexbin Plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", gridsize=15)
plt.suptitle("Hexbin Plot: Total Bill vs Tip")
plt.show()
```

## Regression Plot
A regression plot is used to visualize the linear relationship between two variables with a regression line.

```python
# Regression plot
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot: Total Bill vs Tip")
plt.show()
```

## Regression Residuals Plot
A residuals plot helps visualize the residuals from a linear regression

```python
# Regression Residuals Plot
sns.residplot(x="total_bill", y="tip", data=tips, scatter_kws={"s": 80})
plt.title("Regression Residuals Plot: Total Bill vs Tip")
plt.show()
```

## Facet Grid
A facet grid allows you to create multiple plots for subsets of your data based on a categorical variable.

```python
# Facet grid
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.suptitle("Facet Grid: Scatter Plots for Different Subsets")
plt.show()


```

##  Lineplot with Error Bands
Lineplot can include error bands to visualize uncertainty in the data.



```python
# Lineplot with error bands
sns.lineplot(x="day", y="total_bill", hue="sex", err_style="bars", ci="sd", data=tips)
plt.title("Lineplot with Error Bands: Total Bill by Day with Hue")
plt.show()
```

## Bar Plot with Error Bars
A bar plot with error bars can be useful for showing confidence intervals or standard deviations

```python
# Bar plot with error bars
sns.barplot(x="day", y="total_bill", data=tips, ci="sd")
plt.title("Bar Plot with Error Bars: Total Bill by Day")
plt.show()
```

## Lollipop Plot
A lollipop plot is useful for comparing a single numerical variable across different categories.

```python
# Lollipop plot
sns.set(style="whitegrid")
sns.barplot(x="day", y="total_bill", data=tips, ci=None, palette="Blues_d")
sns.stripplot(x="day", y="total_bill", data=tips, size=8, jitter=True, color="red", linewidth=0.5)
plt.title("Lollipop Plot: Total Bill by Day")
plt.show()
```

## Distribution within Categories
Visualizing the distribution of a numerical variable within different categories using violin plots.

```python
# Violin Plot within Categories
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True)
plt.title("Violin Plot within Categories: Total Bill by Day with Hue")
plt.show()
```

## Time Series Plot

```python
# Time Series Plot
flights = sns.load_dataset("flights")
sns.lineplot(x="year", y="passengers", hue="month", data=flights, marker="o")
plt.title("Time Series Plot: Monthly Passengers Over Years")
plt.show()
```
