# Matplotlib Library


## Relational Charts
### 1- Scatter Plot
A scatter plot displays individual data points on a two-dimensional graph. It is useful for visualizing the relationship between two numerical variables.A scatter plot is used to visualize the relationship between two variables.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12,9, 8, 7]

# Plotting a scatter plot
plt.scatter(x, y, color='red', marker='o')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Example')

# Display the plot
plt.show()
```

### 2- Line chart
A line chart is a common way to represent data points over a continuous interval or time

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]

# Plotting a line chart
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Chart Example')

# Display the plot
plt.show()
```

## Distribution Charts
### 1- Histogram
A histogram represents the distribution of a dataset by dividing it into bins and displaying the frequency of data points in each bin.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)

# Plotting a histogram
plt.hist(data, bins=20, color='green', alpha=0.7)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the plot
plt.show()
```

### 2- Box Plot
A box plot is used to display the distribution and identify outliers in a dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(100, 3)

# Plotting a box plot
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])

# Adding labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Display the plot
plt.show()
```

## Categorical Charts:
### 1- Bar Chart
A bar chart is useful for comparing individual groups or categories.

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C']
values = [25, 40, 30]

# Plotting a bar chart
plt.bar(categories, values, color='skyblue')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Display the plot
plt.show()
```

### 2- Pie Chart
A pie chart is suitable for displaying the contribution of each part to a whole.

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

# Plotting a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue'])

# Adding title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

# Area Chart
An area chart is similar to a line chart but emphasizes the area under the line.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]

# Plotting an area chart
plt.fill_between(x, y, color='skyblue', alpha=0.4)

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Area Chart Example')

# Display the plot
plt.show()
```

# Customized Line Chart
Matplotlib provides a wide range of customization options, including changing colors, styles, and adding annotations

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting a line chart with customization
plt.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='o', markersize=8, label='Sine Wave')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Line Chart')

# Adding grid
plt.grid(True, linestyle='--', alpha=0.7)

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

#  Multiple Subplots
You can create multiple subplots in a single figure using plt.subplots()

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating multiple subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plotting on the first subplot
ax1.plot(x, y1, color='blue', label='Sine')

# Plotting on the second subplot
ax2.plot(x, y2, color='green', label='Cosine')

# Adding labels and title
ax1.set_ylabel('Amplitude')
ax2.set_xlabel('X-axis Label')
ax2.set_ylabel('Amplitude')

# Adding legends
ax1.legend()
ax2.legend()

# Display the plot
plt.show()
```

# Annotations
You can add text annotations and arrows to highlight specific points on a plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting a line chart
plt.plot(x, y, label='Sine Wave')

# Adding an annotation
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Annotation Example')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

# 3D Plot
Matplotlib supports 3D plotting for visualizing three-dimensional data

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Creating a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Adding labels and title
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_zlabel('Z-axis Label')
ax.set_title('3D Surface Plot')

# Display the plot
plt.show()
```

# Subplot Grid
You can create a grid of subplots for more complex visualizations

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a subplot grid
fig, axs = plt.subplots(2, 2)

# Plotting on subplots
axs[0, 0].plot(x, y1, color='blue', label='Sine')
axs[0, 1].plot(x, y2, color='green', label='Cosine')
axs[1, 0].hist(np.random.randn(1000), bins=30, color='orange', alpha=0.7)
axs[1, 1].scatter(x, y1 + y2, color='purple', label='Sum of Sine and Cosine')

# Adding labels and title for the entire figure
fig.suptitle('Subplot Grid Example')

# Display the plot
plt.show()
```

```python
import pandas as pd

# Use double backslashes or a raw string for the file path
file_path = r"D:\DataCamp\Data Scientist with Python\08-Introduction to Data Visualization with Matplotlib\climate_change.csv"

# Read the CSV file
climate_change = pd.read_csv(file_path)

# Display the DataFrame
print(climate_change)
```

```python
# Plotting two time-series together
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"])
ax.plot(climate_change.index, climate_change["relative_temp"])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm) / Relative temperature')
plt.show()
```

```python
# Using twin axes
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"])
ax2.set_ylabel('Relative temperature (Celsius)')
plt.show()
```

```python
# Separating variables by color
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"],color='red')
ax2.set_ylabel('Relative temperature (Celsius)', color='red')
plt.show()
```

```python
# Coloring the ticks
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"],color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax.tick_params('y', colors='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index,climate_change["relative_temp"],color='red')
ax2.set_ylabel('Relative temperature (Celsius)',color='red')
ax2.tick_params('y', colors='red')
plt.show()
```

```python
# A function that plots time-series
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
```

```python
# Using our function
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'],'blue','Time','CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax, climate_change.index,climate_change['relative_temp'],'red','Time','Relative temperature (Celsius)')
plt.show()

```

```python
# Annotation
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time', 'Relative temperature (Celsius)')
ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()

```

```python
mens_rowing = pd.read_csv(r"D:\DataCamp\Data Scientist with Python\08-Introduction to Data Visualization with Matplotlib\summer2016.csv", index_col=0)
mens_rowing.head()
```

```python
# Calculate the mean height for Rowing
mean_height_rowing = mens_rowing.loc[mens_rowing["Sport"] == "Rowing", "Height"].mean()

# Calculate the mean height for gg
mean_height_gg = mens_rowing.loc[mens_rowing["Sport"] == "Gymnastics", "Height"].mean()

# A bar chart
fig, ax = plt.subplots()
ax.bar("Rowing", mean_height_rowing, label="Rowing")
ax.bar("Gymnastics", mean_height_gg, label="Gymnastics")
ax.set_ylabel("Mean Height (cm)")
ax.legend()
plt.show()
```
