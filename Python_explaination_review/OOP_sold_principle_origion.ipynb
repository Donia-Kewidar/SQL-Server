### Inheritance 
is a fundamental concept in Object-Oriented Programming (OOP) that allows a new class to inherit the properties and behaviors of an existing class. It's like a parent-child relationship between classes. The existing class is called the "parent" or "base" class, and the new class is called the "child" or "derived" class.

```python
class ParentClass:
    # Properties and behaviors

class ChildClass(ParentClass):
    # Additional properties and behaviors
    
-----------------------------------------------------------

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def bark(self):
        print(f"{self.name} barks")

class Cat(Animal):
    def meow(self):
        print(f"{self.name} meows")

# Creating instances of the classes
dog = Dog("Buddy")
cat = Cat("Whiskers")

# Accessing inherited methods
dog.speak()  # Output: Buddy makes a sound
cat.speak()  # Output: Whiskers makes a sound

# Accessing methods specific to each class
dog.bark()   # Output: Buddy barks
cat.meow()   # Output: Whiskers meows


-----------------------------------------------------------

class Vehicle:
    def __init__(self, speed):
        self.speed = speed

    def start_engine(self):
        print("Engine started")

class Car(Vehicle):
    def __init__(self, speed, num_doors):
        super().__init__(speed)
        self.num_doors = num_doors

    def honk_horn(self):
        print("Honk honk!")
```

```python
class Food:  # Base Class
    def __init__(self, name, price):
        self.name = name
        self.price = price
        print(f"{self.name} Is Created From Base Class and price is {self.price} ")

    def eat(self):
        print("Eat Method From Base Class")

        
class Apple(Food):  # Derived Class
    def __init__(self, name, price, amount):# it can also overwrite on the base constructor (price+20)
        # Use super() to call the constructor of the parent class (Food)
        super().__init__(name , price) # you must write all the attributes in the constructor (food) 
        self.amount = amount
        self.price = price+20
        print(f"{self.name} Is Created From Derived Class And Price Is {self.price} And Amount Is {self.amount}")

    def get_from_tree(self):
        print("Get From Tree From Derived Class")

# Create an instance of the Apple class with all required arguments
food_two = Apple("Pizza", 150 , 500)

# Call methods from both the base and derived classes
food_two.eat()
food_two.get_from_tree()
```

# Single Responsibility Principle (S.R.P) 
"Do one thing and do it well" (ex: one function(thing) make one job )   
 "There should never be more than one reason for a class to change"    
 A class should have one and only one reason to change, meaning that a class should have only one job.

# Open/Closed Principle (OCP) 

 "The Open/Closed Principle states that software entities (classes, modules, functions, etc.) should be open for extension, but closed for modification(security). "    
 Any new functionality should be done by adding new classes instead of changing existing one
### How to implement OCP 
1-One may achieve this by adding new functionality to derived class (parent)   
2-Or allow client to access the original class with abstract interface

"If you have class B inherits from class A then class A should be replaceable by class B without any changes "
